**Module map.**

| Module | Scope |
|---|---|

**Standards.** IEEE 519-2022 individual and total harmonic limits are applied at the point of common coupling for V less than or equal to 1 kV. Reporting cadence is 10 minutes, consistent with IEC 61000-4-30 Class A aggregation.

## Module 1. Environment, imports, paths, configuration, dataset schemas

This module sets up the runtime environment for the entire pipeline. It performs six tasks:

1. Imports the scientific Python stack used across all subsequent modules.
2. Pins a single global random seed for `random`, `numpy`, and (later) `torch`.
3. Resolves filesystem paths for inputs (the three xlsx files) and outputs (figures, tables, models, logs, interim).
4. Configures matplotlib for thesis-quality figures (300 DPI on save, serif font, A4 text-width sizing, larger fonts than conference style).
5. Defines locked dataset schemas for Cordoba (authoritative, taken from the source repository), Site A, and Site B. Module 2 validates each input file against its schema. Site A and Site B column names use the **real, post-cleanup** strings (with unit suffixes such as `_[Hz]`, `_[V]`, `[Volts]`, `[Amps]`, `[%]`).
6. Defines `CONFIG`, a single nested dictionary that holds every locked methodology constant from Chapter 3 (window length, IEEE 519 thresholds, hyperparameter grids, partition fractions, conformal settings). All later modules read from `CONFIG` rather than hard-coding values.

Heavy framework dependencies (scikit-learn estimators, mapie, statsmodels, torch) are imported in the modules that actually use them, to keep this cell light and to allow the pipeline to fail fast at the relevant module if a dependency is missing.

```python
# Cell 1.5. Locked dataset schemas.
# Cordoba is authoritatively locked from the user-supplied first-row sample (264 columns).
# Site A and Site B column names use the REAL, post-cleanup strings observed in the user's xlsx
# headers; the loader in Module 2 strips stray single quotes and surrounding whitespace before
# any name lookup is performed.
# Expected output: per-site column-count summary, with the Cordoba derived count matching 264.


def _expand_harmonic_columns(template: str, orders: list[int], phases: list[str]) -> list[str]:
    """Expand a column-name template such as 'vmag{order}{phase}' over orders and phases."""
    return [template.format(order=h, phase=p) for h in orders for p in phases]


# ---------------------------------------------------------------------
# Cordoba schema (LOCKED from the user-supplied first-row sample)
# ---------------------------------------------------------------------
CORDOBA_ORDERS: list[int]         = list(range(1, 12))
CORDOBA_VOLTAGE_PHASES: list[str] = ["a", "b", "c"]
CORDOBA_CURRENT_PHASES: list[str] = ["a", "b", "c", "n"]

CORDOBA_SCHEMA: dict[str, Any] = {
    "name":                "cordoba",
    "n_expected_columns":  264,
    "timestamp_col":       "timestamp",
    "timestamp_unit":      "ms",
    "frequency_col":       "f",
    "messageid_col":       "messageid",
    "max_harmonic_order":  11,
    "voltage_phases":      CORDOBA_VOLTAGE_PHASES,
    "current_phases":      CORDOBA_CURRENT_PHASES,
    "vrms_cols":           [f"vrms{p}" for p in CORDOBA_VOLTAGE_PHASES],
    "irms_cols":           [f"irms{p}" for p in CORDOBA_CURRENT_PHASES],
    "thd_v_cols":          [f"thdv{p}" for p in CORDOBA_VOLTAGE_PHASES],
    "thd_i_cols":          [f"thdi{p}" for p in CORDOBA_CURRENT_PHASES],
    "p_cols":              ["pa", "pb", "pc", "ptotal"],
    "q_cols":              ["qa", "qb", "qc", "qtotal"],
    "s_cols":              ["sa", "sb", "sc", "stotal"],
    "pf_cols":             ["tpfa", "tpfb", "tpfc", "tpftotal"],
    "templates": {
        "voltage_magnitude": "vmag{order}{phase}",
        "voltage_phase":     "vph{order}{phase}",
        "voltage_pr":        "vpr{order}{phase}",
        "current_magnitude": "imag{order}{phase}",
        "current_phase":     "iph{order}{phase}",
        "current_pr":        "ipr{order}{phase}",
    },
}
CORDOBA_SCHEMA["vmag_cols"] = _expand_harmonic_columns(
    CORDOBA_SCHEMA["templates"]["voltage_magnitude"], CORDOBA_ORDERS, CORDOBA_VOLTAGE_PHASES)
CORDOBA_SCHEMA["vph_cols"]  = _expand_harmonic_columns(
    CORDOBA_SCHEMA["templates"]["voltage_phase"],     CORDOBA_ORDERS, CORDOBA_VOLTAGE_PHASES)
CORDOBA_SCHEMA["vpr_cols"]  = _expand_harmonic_columns(
    CORDOBA_SCHEMA["templates"]["voltage_pr"],        CORDOBA_ORDERS, CORDOBA_VOLTAGE_PHASES)
CORDOBA_SCHEMA["imag_cols"] = _expand_harmonic_columns(
    CORDOBA_SCHEMA["templates"]["current_magnitude"], CORDOBA_ORDERS, CORDOBA_CURRENT_PHASES)
CORDOBA_SCHEMA["iph_cols"]  = _expand_harmonic_columns(
    CORDOBA_SCHEMA["templates"]["current_phase"],     CORDOBA_ORDERS, CORDOBA_CURRENT_PHASES)
CORDOBA_SCHEMA["ipr_cols"]  = _expand_harmonic_columns(
    CORDOBA_SCHEMA["templates"]["current_pr"],        CORDOBA_ORDERS, CORDOBA_CURRENT_PHASES)

# ---------------------------------------------------------------------
# Site A schema (LOCKED from real headers; loader strips stray quotes first)
# Real cleaned column names use unit suffixes such as F_[Hz], U12_[V], THD12_[%], IL1_[A].
# ---------------------------------------------------------------------
SITE_A_ORDERS: list[int]   = list(range(1, 51))
SITE_A_LL_PAIRS: list[str] = ["12", "23", "31"]
SITE_A_I_PHASES: list[str] = ["1", "2", "3", "N"]   # used in H{order}_I{phase}_[A]; Neutral phase token = "N"

SITE_A_SCHEMA: dict[str, Any] = {
    "name":                "site_a",
    "voltage_form":        "line_to_line",
    "max_harmonic_order":  50,
    "ll_pairs":            SITE_A_LL_PAIRS,
    "current_phases":      SITE_A_I_PHASES,
    "date_col":            "Date",
    "time_col":            "Time",
    "datetime_col":        "Date and Time",
    "datetime_format":     "%Y/%d/%m %H:%M:%S",       # observed format e.g. 2023/21/09 14:50:00
    "frequency_col":       "F_[Hz]",
    "vll_cols":            [f"U{pair}_[V]" for pair in SITE_A_LL_PAIRS],
    "thd_v_cols":          [f"THD{pair}_[%]" for pair in SITE_A_LL_PAIRS],
    "thd_i_cols":          [f"THD_I{p}_[%]" for p in SITE_A_I_PHASES[:3]] + ["THD_IN_[%]"],
    "irms_cols":           [f"IL{p}_[A]" for p in ["1", "2", "3"]] + ["I_Neutral_[A]"],
    "pst_cols":            [f"Pst_U{pair}" for pair in SITE_A_LL_PAIRS],
    "plt_cols":            [f"Plt_U{pair}" for pair in SITE_A_LL_PAIRS],
    "unbalance_v_col":     "Unbalance_-_Voltage_(_Neg./Pos.)_[%]",
    "unbalance_i_col":     "Unbalance_-_Current_(_Neg./Pos.)_[%]",
    "pf_cols":             ["PF_L1", "PF_L2", "PF_L3", "PF_total"],
    "pf_abs_cols":         ["PF_L1_abs", "PF_L2_abs", "PF_L3_abs", "PF_total_abs"],
    "templates": {
        "voltage_magnitude_pct": "H{order}_U{pair}_[%]",
        "current_magnitude_amp": "H{order}_I{phase}_[A]",
    },
    "exclude_orders":      [3, 9],                        # triplens unobservable on L-L wiring
    "phase_v_factor":      1.0 / float(np.sqrt(3.0)),     # V_phase = V_LL / sqrt(3) (balanced approximation)
    "first_row_validated": False,
}
SITE_A_SCHEMA["vmag_pct_cols"] = [
    SITE_A_SCHEMA["templates"]["voltage_magnitude_pct"].format(order=h, pair=p)
    for h in SITE_A_ORDERS for p in SITE_A_LL_PAIRS
]
SITE_A_SCHEMA["imag_amp_cols"] = [
    SITE_A_SCHEMA["templates"]["current_magnitude_amp"].format(order=h, phase=p)
    for h in SITE_A_ORDERS for p in SITE_A_I_PHASES
]

# ---------------------------------------------------------------------
# Site B schema (LOCKED from real headers; column names include unit suffixes)
# Site B has only THD and TDD aggregates; no per-order magnitudes.
# Channels A, B, C are the three phases; D is the fourth meter channel (often neutral or summary feeder).
# ---------------------------------------------------------------------
SITE_B_CHANNELS: list[str] = ["A", "B", "C", "D"]
SITE_B_AGG: list[str]      = ["Max", "Avg", "Min"]

def _b(prefix: str, agg_units: dict[str, str]) -> list[str]:
    """Build Site B column names of the form '{channel}{prefix}{agg}{unit}'."""
    out: list[str] = []
    for c in SITE_B_CHANNELS:
        for a in SITE_B_AGG:
            unit = agg_units.get(a, "")
            out.append(f"{c}{prefix}{a}{unit}")
    return out

SITE_B_SCHEMA: dict[str, Any] = {
    "name":                "site_b",
    "has_per_order":       False,
    "datetime_col":        "Date and Time",
    "time_hours_col":      "Time (hours)",
    "datetime_format":     "%m/%d/%y %I:%M %p",                    # observed format e.g. 5/27/21 3:00 PM
    "channels":            SITE_B_CHANNELS,
    "aggregations":        SITE_B_AGG,
    "primary_aggregation": "Avg",
    "frequency_cols":      [f"TOTFreq{a}[Hz]" for a in SITE_B_AGG],

    "vrms_cols":           _b("Vrms", {"Max": "[Volts]", "Avg": "[Volts]", "Min": "[Volts]"}),
    "irms_cols":           _b("Irms", {"Max": "[Amps]",  "Avg": "[Amps]",  "Min": "[Amps]"}),
    "p_cols":              _b("P(kW)", {"Max": "[kW]",  "Avg": "[kW]",  "Min": "[kW]"}),
    "p_total_cols":        [f"TOTP(kW){a}[kW]"     for a in SITE_B_AGG],
    "s_cols":              _b("S(kVA)", {"Max": "[kVA]", "Avg": "[kVA]", "Min": "[kVA]"}),
    "qfnd_cols":           _b("QFnd(kVAR)", {"Max": "[kVAR]", "Avg": "[kVAR]", "Min": "[kVAR]"}),
    "qfnd_total_cols":     [f"TOTQFnd(kVAR){a}[kVAR]" for a in SITE_B_AGG],
    "pf_cols":             _b("PF", {"Max": "[pu]", "Avg": "[pu]", "Min": "[pu]"}),
    "pf_total_cols":       [f"TOTPF{a}[pu]"           for a in SITE_B_AGG],
    "pf_total_avg_signed": "TOTPFAvg[pu]_not_abs",
    "pf_total_avg":        "TOTPFAvg[pu]",                          # appears at end of file
    "worst_pf_cols":       [f"WorstPF{a}[pu]"         for a in SITE_B_AGG],

    "v_unbalance_cols":    [f"{c}Vunb(DevAvg){a}[%]" for c in ["A", "B", "C"] for a in SITE_B_AGG],
    "v_unbalance_total":   [f"TOTVunb(DevAvg){a}[%]" for a in SITE_B_AGG],

    "thd_v_max_cols":      [f"{c}VThdMax[%]" for c in SITE_B_CHANNELS],
    "thd_v_avg_cols":      [f"{c}VThdAvg[%]" for c in SITE_B_CHANNELS],
    "thd_i_max_cols":      [f"{c}IThdMax[%]" for c in SITE_B_CHANNELS],
    "thd_i_avg_cols":      [f"{c}IThdAvg[%]" for c in SITE_B_CHANNELS],
    "tdd_max_cols":        [f"{c}TDDMax[%]"  for c in SITE_B_CHANNELS],
    "tdd_avg_cols":        [f"{c}TDDAvg[%]"  for c in SITE_B_CHANNELS],

    # Targets locked per Chapter 3: per-phase voltage THD averages (channels A, B, C only).
    "thd_v_target_cols":   ["AVThdAvg[%]", "BVThdAvg[%]", "CVThdAvg[%]"],

    "first_row_validated": False,
}

SCHEMAS: dict[str, dict[str, Any]] = {
    "cordoba": CORDOBA_SCHEMA,
    "site_a":  SITE_A_SCHEMA,
    "site_b":  SITE_B_SCHEMA,
}

# Print column-count summary (Cordoba should match 264 exactly).
_cordoba_count = (
    3                                            # timestamp + f + messageid
    + len(CORDOBA_SCHEMA["vrms_cols"])
    + len(CORDOBA_SCHEMA["irms_cols"])
    + len(CORDOBA_SCHEMA["thd_v_cols"])
    + len(CORDOBA_SCHEMA["thd_i_cols"])
    + len(CORDOBA_SCHEMA["p_cols"])
    + len(CORDOBA_SCHEMA["q_cols"])
    + len(CORDOBA_SCHEMA["s_cols"])
    + len(CORDOBA_SCHEMA["pf_cols"])
    + len(CORDOBA_SCHEMA["vmag_cols"])
    + len(CORDOBA_SCHEMA["vph_cols"])
    + len(CORDOBA_SCHEMA["vpr_cols"])
    + len(CORDOBA_SCHEMA["imag_cols"])
    + len(CORDOBA_SCHEMA["iph_cols"])
    + len(CORDOBA_SCHEMA["ipr_cols"])
)

print(f"{'site':<10} {'derived/known column count':<38} {'expected':<12} {'match':<6}")
print(f"{'cordoba':<10} {_cordoba_count:<38} {CORDOBA_SCHEMA['n_expected_columns']:<12} "
      f"{str(_cordoba_count == CORDOBA_SCHEMA['n_expected_columns']):<6}")
print(f"{'site_a':<10} {'1 datetime + freq + 50*3 V%-orders + 50*4 I[A]-orders + meta (~340)':<38}".replace('340)', '340)'))
print(f"  Site A vmag_pct_cols : {len(SITE_A_SCHEMA['vmag_pct_cols'])} entries (50 orders x 3 LL pairs)")
print(f"  Site A imag_amp_cols : {len(SITE_A_SCHEMA['imag_amp_cols'])} entries (50 orders x 4 channels)")
print(f"{'site_b':<10} aggregates only (no per-order magnitudes); ~120 columns including triplets.")
print()
print(f"Cordoba locked column blocks: {len(CORDOBA_SCHEMA['vmag_cols'])} vmag, "
      f"{len(CORDOBA_SCHEMA['imag_cols'])} imag, "
      f"{len(CORDOBA_SCHEMA['vph_cols'])} vph, "
      f"{len(CORDOBA_SCHEMA['iph_cols'])} iph.")
```

## Module 2. Data ingestion, schema validation, label generation

This module loads the three xlsx files, validates each against its locked schema, parses timestamps, generates IEEE 519-2022 state labels from UTHD, and caches preprocessed frames to parquet under `outputs/interim/`.

Engineering choices baked in:

- **Header cleanup.** Site A xlsx headers carry stray single quotes around many column names and extra surrounding whitespace. The loader normalises every column name with `str.strip().str.strip("'").str.strip()` before any name lookup. The same normaliser is applied uniformly to all three files for consistency, even though Cordoba and Site B headers are already clean.
- **Timestamps.** Cordoba stores epoch milliseconds; Site A stores `YYYY/DD/MM HH:MM:SS` (note: day-month order, not month-day); Site B stores `M/D/YY h:MM AM/PM`. The locked formats live in the schemas in Module 1.
- **Label generation.** Per-phase state labels (Normal / Watch / Exceedance) are derived from UTHD using IEEE 519-2022 thresholds (`tau_W = 6.4%`, `tau_E = 8.0%`). The aggregate worst-state across phases is computed by mapping ordinals to integers; the column is cast to object dtype first, since pandas categorical dtype rejects new categories during fillna and breaks an ordinal map.
- **Schema validation.** For each site, the loader reports observed column count, missing required columns (a fail), and unexpected columns (a warning). It does not modify the data on validation failure; it raises a `ValueError`.
- **Caching.** Each cleaned site frame is written to `outputs/interim/{site}_clean.parquet` for fast reload by Modules 3 onward.

```python
# Cell 2.1. Loader helper utilities.
# Header normalisation, timestamp parsing per site, schema validation report.
# Expected output: no visible output; helpers used by the per-site loaders below.


def normalise_columns(df: pd.DataFrame) -> pd.DataFrame:
    """Strip whitespace and stray single quotes from column names in place."""
    df.columns = (
        pd.Index(df.columns.astype(str))
          .str.strip()
          .str.strip("'")
          .str.strip()
    )
    # Defensive de-duplication for spreadsheet artefacts (e.g. Site B shows
    # 'TOTPFAvg[pu]' both early and at the end of the file).
    if df.columns.duplicated().any():
        df = df.loc[:, ~df.columns.duplicated(keep="first")]
    return df


def parse_timestamp_ms(series: pd.Series) -> pd.DatetimeIndex:
    """Convert a pandas Series of epoch milliseconds to a UTC DatetimeIndex."""
    numeric = pd.to_numeric(series, errors="coerce")
    return pd.to_datetime(numeric, unit="ms", utc=True, errors="coerce")


def parse_timestamp_str(series: pd.Series, fmt: str) -> pd.DatetimeIndex:
    """Parse a string-formatted timestamp series with a fixed strptime format."""
    return pd.to_datetime(series.astype(str).str.strip(), format=fmt, errors="coerce")


def validate_schema(
    df: pd.DataFrame,
    required: list[str],
    site_name: str,
    *,
    strict: bool = True,
) -> dict[str, Any]:
    """Compare the cleaned DataFrame columns against a list of required column names.

    Returns a dict with keys 'observed', 'missing', 'matched', 'unexpected'.
    Raises ValueError if `strict` is True and any required columns are missing.
    """
    observed = list(df.columns)
    obs_set  = set(observed)
    req_set  = set(required)
    missing    = sorted(req_set - obs_set)
    matched    = sorted(req_set & obs_set)
    unexpected = sorted(obs_set - req_set)
    report = {
        "site":         site_name,
        "observed":     len(observed),
        "required":     len(required),
        "matched":      len(matched),
        "missing":      missing,
        "unexpected":   unexpected,
    }
    if missing and strict:
        raise ValueError(
            f"[{site_name}] missing {len(missing)} required columns: "
            f"{missing[:8]}{'...' if len(missing) > 8 else ''}"
        )
    return report
```

```python
# Cell 2.2. IEEE 519-2022 label helpers.
# Per-phase ternary label from UTHD; aggregate worst-state across phases with
# the categorical-to-object cast fix.
# Expected output: no visible output; helpers used by the per-site loaders.

# Ordinal mapping: Normal < Watch < Exceedance.
STATES: list[str] = ["Normal", "Watch", "Exceedance"]
STATE_TO_ORD: dict[str, int] = {"Normal": 0, "Watch": 1, "Exceedance": 2}
ORD_TO_STATE: dict[int, str] = {v: k for k, v in STATE_TO_ORD.items()}


def label_uthd(uthd_pct: pd.Series, *, tau_W: float, tau_E: float) -> pd.Series:
    """Return a ternary-state Series from a UTHD percentage Series.

    States are 'Normal' (UTHD < tau_W), 'Watch' (tau_W <= UTHD < tau_E),
    and 'Exceedance' (UTHD >= tau_E). NaN UTHD values yield NaN states.
    """
    out = pd.Series(np.full(len(uthd_pct), np.nan, dtype=object), index=uthd_pct.index)
    valid = uthd_pct.notna()
    vals  = uthd_pct[valid].astype(float).to_numpy()
    states = np.where(
        vals >= tau_E, "Exceedance",
        np.where(vals >= tau_W, "Watch", "Normal"),
    )
    out.loc[valid] = states
    return out


def aggregate_worst_state(state_df: pd.DataFrame) -> pd.Series:
    """Aggregate per-phase state columns into a single worst-state column.

    Worst here means the ordinally largest state across phases at each row.
    The cast to object before mapping is the documented fix: pandas categorical
    dtype rejects the integer 0 produced by fillna for unmapped (NaN) values.
    """
    # Cast to object so that .map can introduce NaN cleanly (categorical refuses).
    obj_df = state_df.astype(object)
    ord_df = obj_df.apply(lambda s: s.map(STATE_TO_ORD)).fillna(0).astype(int)
    worst_ord = ord_df.max(axis=1)
    return worst_ord.map(ORD_TO_STATE).astype("string")
```

```python
# Cell 2.3. Cordoba loader.
# Reads the xlsx, cleans headers, converts the epoch-ms timestamp, validates
# the 264-column schema, generates per-phase and aggregate UTHD state labels,
# and caches the result to outputs/interim/cordoba_clean.parquet.
# Expected output: shape, time range, schema match, label class distribution.


def load_cordoba(path: Path, schema: dict[str, Any]) -> tuple[pd.DataFrame, dict[str, Any]]:
    """Load and clean the Cordoba dataset; return (frame, validation_report)."""
    df = pd.read_excel(path, engine="openpyxl")
    df = normalise_columns(df)

    # Required columns: timestamp + frequency + messageid + a representative subset.
    required = (
        [schema["timestamp_col"], schema["frequency_col"], schema["messageid_col"]]
        + schema["vrms_cols"] + schema["irms_cols"]
        + schema["thd_v_cols"] + schema["thd_i_cols"]
        + schema["p_cols"] + schema["q_cols"] + schema["s_cols"] + schema["pf_cols"]
        + schema["vmag_cols"] + schema["vph_cols"] + schema["vpr_cols"]
        + schema["imag_cols"] + schema["iph_cols"] + schema["ipr_cols"]
    )
    report = validate_schema(df, required, site_name="cordoba", strict=True)

    # Timestamp: epoch milliseconds, UTC. Set as DatetimeIndex 'ts'.
    ts = parse_timestamp_ms(df[schema["timestamp_col"]])
    df = df.assign(ts=ts).set_index("ts").sort_index()

    # UTHD per phase already present as thdva/thdvb/thdvc.
    tau_W = CONFIG["ieee519"]["tau_W_uthd_pct"]
    tau_E = CONFIG["ieee519"]["tau_E_uthd_pct"]
    for phase, col in zip(schema["voltage_phases"], schema["thd_v_cols"]):
        df[f"state_{phase}"] = label_uthd(df[col], tau_W=tau_W, tau_E=tau_E)
    df["state_worst"] = aggregate_worst_state(
        df[[f"state_{p}" for p in schema["voltage_phases"]]]
    )
    return df, report


_cordoba_df, _cordoba_report = load_cordoba(CORDOBA_XLSX, CORDOBA_SCHEMA)
print(f"Cordoba shape       : {_cordoba_df.shape}")
print(f"Cordoba time range  : {_cordoba_df.index.min()} -> {_cordoba_df.index.max()}")
print(f"Cordoba columns     : observed={_cordoba_report['observed']}, "
      f"required={_cordoba_report['required']}, matched={_cordoba_report['matched']}, "
      f"unexpected={len(_cordoba_report['unexpected'])}")
print("\nCordoba state distribution (worst across phases):")
print(_cordoba_df["state_worst"].value_counts(dropna=False).to_string())

# Cache to parquet.
_cordoba_cache = INTERIM_DIR / "cordoba_clean.parquet"
_cordoba_df.to_parquet(_cordoba_cache)
print(f"\nCached: {_cordoba_cache}")
```

```python
# Cell 2.4. Site A loader.
# Strips stray quotes from headers, parses the combined 'Date and Time' column,
# verifies the line-to-line voltage and current per-order column blocks, computes
# UTHD per phase from H{order}_U{pair}_[%] columns (excluding triplens 3 and 9),
# and generates state labels.
# Expected output: shape, time range, schema match, label distribution.


def compute_site_a_uthd(df: pd.DataFrame, schema: dict[str, Any]) -> pd.DataFrame:
    """Recompute UTHD per L-L pair from H{h}_U{pair}_[%] columns, excluding triplens.

    The native THD12_[%] etc. columns are also retained, but the recomputed value
    UTHD_recomputed_{pair} is what the detection logic uses (the exclusion of
    triplens makes Site A's UTHD comparable across the cross-site protocol).
    """
    excl = set(schema["exclude_orders"])
    out = df.copy()
    for pair in schema["ll_pairs"]:
        per_order_cols = [
            f"H{h}_U{pair}_[%]" for h in range(2, schema["max_harmonic_order"] + 1)
            if h not in excl and f"H{h}_U{pair}_[%]" in df.columns
        ]
        if per_order_cols:
            sq_sum = (df[per_order_cols].astype(float) ** 2).sum(axis=1)
            out[f"UTHD_recomputed_{pair}"] = np.sqrt(sq_sum)
        else:
            out[f"UTHD_recomputed_{pair}"] = np.nan
    return out


def load_site_a(path: Path, schema: dict[str, Any]) -> tuple[pd.DataFrame, dict[str, Any]]:
    """Load and clean the Site A dataset; return (frame, validation_report)."""
    df = pd.read_excel(path, engine="openpyxl")
    df = normalise_columns(df)

    # Minimum required: combined datetime, frequency, the three L-L THD columns,
    # the four current-RMS columns, and the unbalance columns.
    required = (
        [schema["datetime_col"], schema["frequency_col"]]
        + schema["vll_cols"] + schema["thd_v_cols"]
        + schema["irms_cols"] + schema["thd_i_cols"]
        + [schema["unbalance_v_col"], schema["unbalance_i_col"]]
    )
    report = validate_schema(df, required, site_name="site_a", strict=True)

    # Per-order block coverage check (warn-only).
    missing_v_orders = [c for c in schema["vmag_pct_cols"] if c not in df.columns]
    missing_i_orders = [c for c in schema["imag_amp_cols"] if c not in df.columns]
    report["missing_per_order_voltage_count"] = len(missing_v_orders)
    report["missing_per_order_current_count"] = len(missing_i_orders)

    # Timestamp.
    ts = parse_timestamp_str(df[schema["datetime_col"]], schema["datetime_format"])
    df = df.assign(ts=ts).set_index("ts").sort_index()

    # UTHD recomputed from per-order columns (triplens excluded for L-L comparability).
    df = compute_site_a_uthd(df, schema)

    # State labels using recomputed UTHD per L-L pair.
    tau_W = CONFIG["ieee519"]["tau_W_uthd_pct"]
    tau_E = CONFIG["ieee519"]["tau_E_uthd_pct"]
    for pair in schema["ll_pairs"]:
        df[f"state_{pair}"] = label_uthd(
            df[f"UTHD_recomputed_{pair}"], tau_W=tau_W, tau_E=tau_E
        )
    df["state_worst"] = aggregate_worst_state(
        df[[f"state_{p}" for p in schema["ll_pairs"]]]
    )

    return df, report


_site_a_df, _site_a_report = load_site_a(SITE_A_XLSX, SITE_A_SCHEMA)
print(f"Site A shape        : {_site_a_df.shape}")
print(f"Site A time range   : {_site_a_df.index.min()} -> {_site_a_df.index.max()}")
print(f"Site A columns      : observed={_site_a_report['observed']}, "
      f"required={_site_a_report['required']}, matched={_site_a_report['matched']}, "
      f"missing_per_order_V={_site_a_report['missing_per_order_voltage_count']}, "
      f"missing_per_order_I={_site_a_report['missing_per_order_current_count']}")
print("\nSite A state distribution (worst across L-L pairs):")
print(_site_a_df["state_worst"].value_counts(dropna=False).to_string())

_site_a_cache = INTERIM_DIR / "site_a_clean.parquet"
_site_a_df.to_parquet(_site_a_cache)
print(f"\nCached: {_site_a_cache}")
```

```python
# Cell 2.5. Site B loader.
# Cleans headers (de-duplicates 'TOTPFAvg[pu]' when it appears twice), parses
# the Date and Time string, validates aggregate-only schema, generates state
# labels using AVThdAvg[%], BVThdAvg[%], CVThdAvg[%] (channels A/B/C only).
# Expected output: shape, time range, schema match, label distribution.


def load_site_b(path: Path, schema: dict[str, Any]) -> tuple[pd.DataFrame, dict[str, Any]]:
    """Load and clean the Site B dataset; return (frame, validation_report)."""
    df = pd.read_excel(path, engine="openpyxl")
    df = normalise_columns(df)

    # Required: datetime, frequency aggregates, per-channel V/I rms aggregates,
    # voltage THD avg targets, current THD avgs, TDD avgs.
    required = (
        [schema["datetime_col"]]
        + schema["frequency_cols"]
        + schema["vrms_cols"] + schema["irms_cols"]
        + schema["thd_v_avg_cols"] + schema["thd_i_avg_cols"]
        + schema["tdd_avg_cols"]
        + schema["thd_v_target_cols"]
    )
    # Site B is treated more leniently than Cordoba: warn on missing rather than fail
    # if and only if the missing columns are non-essential aggregations.
    report = validate_schema(df, required, site_name="site_b", strict=True)

    # Timestamp.
    ts = parse_timestamp_str(df[schema["datetime_col"]], schema["datetime_format"])
    df = df.assign(ts=ts).set_index("ts").sort_index()

    # Per-channel state labels using A/B/C voltage THD averages (D excluded from targets).
    tau_W = CONFIG["ieee519"]["tau_W_uthd_pct"]
    tau_E = CONFIG["ieee519"]["tau_E_uthd_pct"]
    target_phases = ["A", "B", "C"]
    for ch, col in zip(target_phases, schema["thd_v_target_cols"]):
        df[f"state_{ch}"] = label_uthd(df[col], tau_W=tau_W, tau_E=tau_E)
    df["state_worst"] = aggregate_worst_state(
        df[[f"state_{ch}" for ch in target_phases]]
    )

    return df, report


_site_b_df, _site_b_report = load_site_b(SITE_B_XLSX, SITE_B_SCHEMA)
print(f"Site B shape        : {_site_b_df.shape}")
print(f"Site B time range   : {_site_b_df.index.min()} -> {_site_b_df.index.max()}")
print(f"Site B columns      : observed={_site_b_report['observed']}, "
      f"required={_site_b_report['required']}, matched={_site_b_report['matched']}")
print("\nSite B state distribution (worst across A/B/C):")
print(_site_b_df["state_worst"].value_counts(dropna=False).to_string())

_site_b_cache = INTERIM_DIR / "site_b_clean.parquet"
_site_b_df.to_parquet(_site_b_cache)
print(f"\nCached: {_site_b_cache}")
```

## Module 3. Preprocessing: cadence, harmonisation, partitioning, frozen scaler

This module turns each cleaned site frame into a modelling-ready frame and a chronologically partitioned canonical common feature matrix. It performs five tasks:

1. **Cadence diagnostics and reindexing.** Drop exact-duplicate timestamps (the dominant component of the 16% off-cadence flagged in the prior Cordoba run), report the empirical interval distribution, then reindex each site to a strict 10-min DatetimeIndex over its observed span. Single-step gaps (one missing 10-min sample) are filled by time-based linear interpolation on numeric columns; longer gaps remain NaN and are excluded from any feature window that overlaps them.
2. **Canonical 18-column harmonisation.** Each site is mapped onto the common feature schema: `supply_freq`, `thd_v_{a,b,c}`, `vrms_{a,b,c}`, `irms_{a,b,c}`, `p_{a,b,c}`, `q_{a,b,c}`, `v_unbalance`, `i_unbalance`. Site A is converted from line-to-line to phase-equivalent voltage by the balanced approximation `V_phase = V_LL / sqrt(3)`; phase active and reactive powers are derived from the per-phase RMS values and signed power factor.
3. **Per-phase derived UTHD on the harmonised frame.** Each site's modelling frame keeps a single `uthd_pct_{a,b,c}` triple (drawn from the native or recomputed THD columns) so detection logic in Module 5 has a uniform entry point.
4. **Chronological 70/15/15 partition labels.** A `partition` column with values `train`, `calib`, `test` is added to each site's frame. No information from `calib` or `test` is allowed into any fit step downstream.
5. **Cordoba-frozen common-feature scaler.** A `StandardScaler` is fit on Cordoba's training partition over the canonical 18 columns and pickled. Module 10 reloads it for cross-site transfer; per-site scalers fit in Module 4 on local training partitions are separate.

Limitations explicitly noted:

- Site A line-to-line THD is preserved as the per-phase THD proxy. THD is dimensionless and does not transform under the LL/phase change; using LL THD as the phase THD is the documented approximation. Triplen orders 3 and 9 are excluded from Site A's recomputed UTHD because L-L wiring suppresses zero-sequence triplens.
- Site A reactive power is reconstructed as `Q = S sqrt(1 - PF^2)` with the sign of `Q` lost when `PF_signed` and `PF_abs` agree. This is acceptable for the cross-site transfer common subset.
- Site B has no native current unbalance column. It is computed from the maximum per-phase RMS deviation from the three-phase mean.

```python
# Cell 3.1. Cadence diagnostics and strict-grid reindex.
# Drops duplicate timestamps, reports the inter-sample interval distribution, then reindexes
# the frame to a strict 10-min grid. Single-step gaps are interpolated; longer gaps are NaN.
# Expected output: per-site interval summary printed; figure of off-cadence count.


def cadence_diagnostics(df: pd.DataFrame, *, label: str) -> pd.DataFrame:
    """Report duplicate count and inter-sample interval distribution.

    Returns a one-row summary DataFrame; also prints a brief table.
    """
    n_total       = len(df)
    n_duplicates  = int(df.index.duplicated(keep="first").sum())
    diffs         = df.index.to_series().diff().dropna()
    n_off_10min   = int((diffs != pd.Timedelta(minutes=10)).sum())
    summary = pd.DataFrame(
        {
            "site":           [label],
            "rows_total":     [n_total],
            "rows_duplicate": [n_duplicates],
            "off_10min_count":[n_off_10min],
            "off_10min_pct":  [100.0 * n_off_10min / max(1, len(diffs))],
            "min_dt":         [str(diffs.min()) if len(diffs) else "n/a"],
            "max_dt":         [str(diffs.max()) if len(diffs) else "n/a"],
            "median_dt":      [str(diffs.median()) if len(diffs) else "n/a"],
        }
    )
    print(summary.to_string(index=False))
    return summary


def reindex_strict_10min(
    df: pd.DataFrame,
    *,
    interp_max_steps: int = 1,
    label: str = "",
) -> tuple[pd.DataFrame, dict[str, int]]:
    """Drop duplicate timestamps then reindex to a strict 10-min grid.

    Numeric columns are time-interpolated only across gaps of at most
    `interp_max_steps` consecutive missing samples (default: one 10-min step,
    so 20-min gaps and longer remain NaN). Object / string / category columns
    are left as NaN at gap rows; downstream code re-derives state labels.
    """
    # Drop duplicate timestamps.
    n_dupes = int(df.index.duplicated(keep="first").sum())
    df = df[~df.index.duplicated(keep="first")]

    # Reindex to strict grid.
    full_idx = pd.date_range(
        start=df.index.min(), end=df.index.max(), freq="10min", tz=df.index.tz
    )
    n_added = len(full_idx) - len(df)
    out = df.reindex(full_idx)
    out.index.name = "ts"

    # Interpolate numeric columns across short gaps only.
    numeric_cols = out.select_dtypes(include=[np.number]).columns
    if len(numeric_cols) and interp_max_steps > 0:
        out[numeric_cols] = out[numeric_cols].interpolate(
            method="time", limit=interp_max_steps, limit_direction="forward"
        )

    info = {
        "duplicates_dropped": n_dupes,
        "rows_added_by_grid": n_added,
        "rows_after_grid":    len(out),
    }
    print(f"[{label}] duplicates_dropped={info['duplicates_dropped']}, "
          f"rows_added_by_grid={info['rows_added_by_grid']}, "
          f"rows_after_grid={info['rows_after_grid']}")
    return out, info


print("Cadence diagnostics (raw, post Module 2):")
_diag_cordoba = cadence_diagnostics(_cordoba_df, label="cordoba")
_diag_site_a  = cadence_diagnostics(_site_a_df,  label="site_a")
_diag_site_b  = cadence_diagnostics(_site_b_df,  label="site_b")

print("\nStrict 10-min reindex with single-step interpolation:")
_cordoba_grid, _cordoba_grid_info = reindex_strict_10min(_cordoba_df, label="cordoba")
_site_a_grid,  _site_a_grid_info  = reindex_strict_10min(_site_a_df,  label="site_a")
_site_b_grid,  _site_b_grid_info  = reindex_strict_10min(_site_b_df,  label="site_b")
```

```python
# Cell 3.2. Re-derive state labels on the reindexed grid.
# After interpolation, UTHD on freshly filled rows is finite; rows still NaN remain unlabelled.
# Expected output: updated state distribution printed per site.


def relabel_states(
    df: pd.DataFrame,
    uthd_cols: list[str],
    state_keys: list[str],
    *,
    tau_W: float,
    tau_E: float,
) -> pd.DataFrame:
    """Recompute per-phase state columns and an aggregate worst-state column.

    `uthd_cols` and `state_keys` must be aligned: state_keys[i] is the suffix
    used for the i-th UTHD column (e.g. 'a' for thdva, '12' for THD12_[%], 'A'
    for AVThdAvg[%]).
    """
    out = df.copy()
    for col, key in zip(uthd_cols, state_keys):
        out[f"state_{key}"] = label_uthd(out[col], tau_W=tau_W, tau_E=tau_E)
    out["state_worst"] = aggregate_worst_state(
        out[[f"state_{k}" for k in state_keys]]
    )
    return out


_tw = CONFIG["ieee519"]["tau_W_uthd_pct"]
_te = CONFIG["ieee519"]["tau_E_uthd_pct"]

_cordoba_grid = relabel_states(
    _cordoba_grid,
    uthd_cols=CORDOBA_SCHEMA["thd_v_cols"],
    state_keys=CORDOBA_VOLTAGE_PHASES,
    tau_W=_tw, tau_E=_te,
)
_site_a_grid = relabel_states(
    _site_a_grid,
    uthd_cols=[f"UTHD_recomputed_{p}" for p in SITE_A_LL_PAIRS],
    state_keys=SITE_A_LL_PAIRS,
    tau_W=_tw, tau_E=_te,
)
_site_b_grid = relabel_states(
    _site_b_grid,
    uthd_cols=SITE_B_SCHEMA["thd_v_target_cols"],
    state_keys=["A", "B", "C"],
    tau_W=_tw, tau_E=_te,
)

for _name, _df in [("cordoba", _cordoba_grid), ("site_a", _site_a_grid), ("site_b", _site_b_grid)]:
    print(f"[{_name}] worst-state distribution after reindex+relabel:")
    print(_df["state_worst"].value_counts(dropna=False).to_string())
    print()
```

```python
# Cell 3.3. Three-phase unbalance helpers.
# When a native unbalance column is unavailable, compute the simple maximum-deviation
# unbalance ratio: max(|X_p - X_avg|) / X_avg * 100% over the three phases.
# Expected output: no visible output; helpers used by the per-site harmonisers.


def max_deviation_unbalance(df: pd.DataFrame, cols: list[str]) -> pd.Series:
    """Return the maximum-deviation three-phase unbalance percentage.

    For each row, computes mean across `cols`, then 100 * max|x_p - mean| / mean.
    Returns NaN where mean is non-positive or all values are NaN.
    """
    vals = df[cols].astype(float)
    mu = vals.mean(axis=1)
    safe_mu = mu.where(mu > 0)
    dev = vals.sub(mu, axis=0).abs().max(axis=1)
    return 100.0 * dev / safe_mu
```

```python
# Cell 3.4. Cordoba canonical 18-column common feature builder.
# Maps native columns to the locked common schema; computes V/I unbalance from per-phase RMS.
# Expected output: shape and head of the resulting canonical frame.

CANONICAL_COLS: list[str] = [
    "supply_freq",
    "thd_v_a", "thd_v_b", "thd_v_c",
    "vrms_a",  "vrms_b",  "vrms_c",
    "irms_a",  "irms_b",  "irms_c",
    "p_a",     "p_b",     "p_c",
    "q_a",     "q_b",     "q_c",
    "v_unbalance",
    "i_unbalance",
]
assert len(CANONICAL_COLS) == 18, "Common feature schema is locked at 18 columns."


def build_common_features_cordoba(df: pd.DataFrame) -> pd.DataFrame:
    """Build the 18-column canonical common feature frame for Cordoba."""
    out = pd.DataFrame(index=df.index)
    out["supply_freq"] = df["f"]
    out["thd_v_a"], out["thd_v_b"], out["thd_v_c"] = df["thdva"], df["thdvb"], df["thdvc"]
    out["vrms_a"],  out["vrms_b"],  out["vrms_c"]  = df["vrmsa"],  df["vrmsb"],  df["vrmsc"]
    out["irms_a"],  out["irms_b"],  out["irms_c"]  = df["irmsa"],  df["irmsb"],  df["irmsc"]
    out["p_a"], out["p_b"], out["p_c"] = df["pa"], df["pb"], df["pc"]
    out["q_a"], out["q_b"], out["q_c"] = df["qa"], df["qb"], df["qc"]
    out["v_unbalance"] = max_deviation_unbalance(df, ["vrmsa", "vrmsb", "vrmsc"])
    out["i_unbalance"] = max_deviation_unbalance(df, ["irmsa", "irmsb", "irmsc"])
    out = out.reindex(columns=CANONICAL_COLS)
    return out


_cordoba_common = build_common_features_cordoba(_cordoba_grid)
print(f"Cordoba common-feature frame shape: {_cordoba_common.shape} (expect cols=18).")
print(_cordoba_common.head(3).to_string())
```

```python
# Cell 3.5. Site A canonical 18-column common feature builder.
# Converts L-L to phase-equivalent voltages; reconstructs P/Q from V, I, and signed PF.
# Expected output: shape and head of the resulting canonical frame.


def build_common_features_site_a(df: pd.DataFrame, schema: dict[str, Any]) -> pd.DataFrame:
    """Build the canonical common feature frame for Site A.

    Phase mapping:
        a <- pair 12 (RMS U12 / sqrt(3)),  current IL1, PF_L1
        b <- pair 23 (RMS U23 / sqrt(3)),  current IL2, PF_L2
        c <- pair 31 (RMS U31 / sqrt(3)),  current IL3, PF_L3
    """
    factor = schema["phase_v_factor"]
    out = pd.DataFrame(index=df.index)
    out["supply_freq"] = df[schema["frequency_col"]]

    # THD passes through (LL THD used as phase THD proxy under the documented approximation).
    out["thd_v_a"], out["thd_v_b"], out["thd_v_c"] = df["THD12_[%]"], df["THD23_[%]"], df["THD31_[%]"]

    # Phase-equivalent voltages.
    out["vrms_a"] = df["U12_[V]"].astype(float) * factor
    out["vrms_b"] = df["U23_[V]"].astype(float) * factor
    out["vrms_c"] = df["U31_[V]"].astype(float) * factor

    # Per-phase line currents.
    out["irms_a"] = df["IL1_[A]"].astype(float)
    out["irms_b"] = df["IL2_[A]"].astype(float)
    out["irms_c"] = df["IL3_[A]"].astype(float)

    # Reconstruct P and Q from V, I, and PF.
    pf_a = df["PF_L1"].astype(float)
    pf_b = df["PF_L2"].astype(float)
    pf_c = df["PF_L3"].astype(float)
    s_a = out["vrms_a"] * out["irms_a"]
    s_b = out["vrms_b"] * out["irms_b"]
    s_c = out["vrms_c"] * out["irms_c"]
    out["p_a"] = s_a * pf_a
    out["p_b"] = s_b * pf_b
    out["p_c"] = s_c * pf_c
    out["q_a"] = s_a * np.sqrt(np.clip(1.0 - pf_a ** 2, a_min=0.0, a_max=None))
    out["q_b"] = s_b * np.sqrt(np.clip(1.0 - pf_b ** 2, a_min=0.0, a_max=None))
    out["q_c"] = s_c * np.sqrt(np.clip(1.0 - pf_c ** 2, a_min=0.0, a_max=None))

    # Native unbalance (negative / positive sequence ratio percent).
    out["v_unbalance"] = df[schema["unbalance_v_col"]].astype(float)
    out["i_unbalance"] = df[schema["unbalance_i_col"]].astype(float)

    out = out.reindex(columns=CANONICAL_COLS)
    return out


_site_a_common = build_common_features_site_a(_site_a_grid, SITE_A_SCHEMA)
print(f"Site A common-feature frame shape: {_site_a_common.shape} (expect cols=18).")
print(_site_a_common.head(3).to_string())
```

```python
# Cell 3.6. Site B canonical 18-column common feature builder.
# Uses Avg aggregations across the board; current unbalance derived from per-phase RMS.
# Expected output: shape and head of the resulting canonical frame.


def build_common_features_site_b(df: pd.DataFrame, schema: dict[str, Any]) -> pd.DataFrame:
    """Build the canonical common feature frame for Site B (channels A/B/C only)."""
    out = pd.DataFrame(index=df.index)
    out["supply_freq"] = df["TOTFreqAvg[Hz]"].astype(float)
    out["thd_v_a"], out["thd_v_b"], out["thd_v_c"] = (
        df["AVThdAvg[%]"].astype(float),
        df["BVThdAvg[%]"].astype(float),
        df["CVThdAvg[%]"].astype(float),
    )
    out["vrms_a"], out["vrms_b"], out["vrms_c"] = (
        df["AVrmsAvg[Volts]"].astype(float),
        df["BVrmsAvg[Volts]"].astype(float),
        df["CVrmsAvg[Volts]"].astype(float),
    )
    out["irms_a"], out["irms_b"], out["irms_c"] = (
        df["AIrmsAvg[Amps]"].astype(float),
        df["BIrmsAvg[Amps]"].astype(float),
        df["CIrmsAvg[Amps]"].astype(float),
    )
    out["p_a"], out["p_b"], out["p_c"] = (
        df["AP(kW)Avg[kW]"].astype(float),
        df["BP(kW)Avg[kW]"].astype(float),
        df["CP(kW)Avg[kW]"].astype(float),
    )
    out["q_a"], out["q_b"], out["q_c"] = (
        df["AQFnd(kVAR)Avg[kVAR]"].astype(float),
        df["BQFnd(kVAR)Avg[kVAR]"].astype(float),
        df["CQFnd(kVAR)Avg[kVAR]"].astype(float),
    )
    out["v_unbalance"] = df["TOTVunb(DevAvg)Avg[%]"].astype(float)
    out["i_unbalance"] = max_deviation_unbalance(
        df, ["AIrmsAvg[Amps]", "BIrmsAvg[Amps]", "CIrmsAvg[Amps]"]
    )
    out = out.reindex(columns=CANONICAL_COLS)
    return out


_site_b_common = build_common_features_site_b(_site_b_grid, SITE_B_SCHEMA)
print(f"Site B common-feature frame shape: {_site_b_common.shape} (expect cols=18).")
print(_site_b_common.head(3).to_string())
```

```python
# Cell 3.7. Chronological 70/15/15 partition labels.
# Adds a 'partition' column to the canonical frames; downstream fits use train rows only.
# Expected output: per-site partition counts.


def add_partition_column(df: pd.DataFrame, *, train_frac: float, calib_frac: float) -> pd.DataFrame:
    """Append a 'partition' column with values 'train', 'calib', 'test' chronologically.

    Rows are NOT shuffled. The function does not alter the index or any other columns.
    """
    n = len(df)
    n_train = int(round(train_frac * n))
    n_calib = int(round(calib_frac * n))
    labels = np.empty(n, dtype=object)
    labels[:n_train]                    = "train"
    labels[n_train:n_train + n_calib]   = "calib"
    labels[n_train + n_calib:]          = "test"
    out = df.copy()
    out["partition"] = labels
    return out


_train_frac = CONFIG["partitions"]["train_frac"]
_calib_frac = CONFIG["partitions"]["calib_frac"]

_cordoba_common = add_partition_column(_cordoba_common, train_frac=_train_frac, calib_frac=_calib_frac)
_site_a_common  = add_partition_column(_site_a_common,  train_frac=_train_frac, calib_frac=_calib_frac)
_site_b_common  = add_partition_column(_site_b_common,  train_frac=_train_frac, calib_frac=_calib_frac)

for _name, _df in [("cordoba", _cordoba_common), ("site_a", _site_a_common), ("site_b", _site_b_common)]:
    counts = _df["partition"].value_counts().reindex(["train", "calib", "test"]).fillna(0).astype(int)
    print(f"[{_name}] partitions  train={counts['train']}  calib={counts['calib']}  test={counts['test']}")
```

## Module 4. Feature engineering: sliding-window, deltas, spectrum, complex phasors

This module builds the modelling feature matrices. Two products are produced:

- **Common derived features** for every site, identical column names across sites, used both for per-site modelling and for the cross-site frozen-model transfer in Module 10. Built from the canonical 18-column common feature frame produced in Module 3.
- **Cordoba enrichment features** added on top of the common derived set. These features depend on per-order phasor data (orders 1 through 11 for both voltage and current, plus per-order phase angles and prevailing-phasor ratios) that are present at Cordoba but missing or unobservable at Sites A and B. The enrichment block is reserved for Cordoba's primary forecasting run and for the ablation study in Module 9.

Mathematics, locked from Chapter 3:

- Sliding-window length `W = 6` (one hour at 10-min cadence). For each base feature `x`:
  - mean `mu_x(t)`, standard deviation `sigma_x(t)`, max, min over the last `W` samples.
  - ramp rate `r_x(t) = (x(t) - x(t - W + 1)) / ((W - 1) * dt)` with `dt = 600 s`.
- Delta features at lags `ell in {1, 3, 6}`: `Delta_ell x(t) = x(t) - x(t - ell)`.
- Spectrum-shape ratios on Cordoba per voltage and current phase, over orders 2 through 11:
  - `R_oe = sum_{h in odd} V_h^2 / (sum_{h in even} V_h^2 + eps)` with `odd = {3,5,7,9,11}`, `even = {2,4,6,8,10}`.
  - `R_tri = (V_3^2 + V_9^2) / (sum_{h=2}^{11} V_h^2 + eps)`.
  - `C_U = sum_{h=2}^{11} h * V_h^2 / (sum_{h=2}^{11} V_h^2 + eps)`.
- Complex phasor features on Cordoba at target orders `{3, 5, 7, 9, 11}`:
  - Per-order complex impedance `Z_h = U_h e^{j phi_h^U} / I_h e^{j phi_h^I}` (real and imaginary parts).
  - Per-order real and reactive harmonic power `P_h + j Q_h = U_h conj(I_h)`.
  - Circular embedding of voltage and current phase angles via `sin` and `cos` (degrees converted to radians).
  - Prevailing-phasor confidence ratios `vpr_h` and `ipr_h` passed through as features.

`eps = 1e-6` throughout. Rows where the sliding-window descriptor is undefined (the first `W - 1` rows after each contiguous segment) carry NaN and are excluded by drop in the modelling step.

```python
# Cell 4.1. Feature-engineering helper functions.
# Sliding-window descriptors and lagged deltas over a list of base columns.
# Expected output: no visible output; helpers used by the per-site builders.


def sliding_window_features(df: pd.DataFrame, cols: list[str], *, W: int) -> pd.DataFrame:
    """Build mean / std / max / min / ramp features over a length-W trailing window.

    Standard deviation uses the population definition (`ddof=0`) for numerical
    stability when the window is fully populated. Ramp rate has units of base / second.
    """
    if W < 2:
        raise ValueError("W must be >= 2 for ramp computation.")
    out = pd.DataFrame(index=df.index)
    rolling = df[cols].rolling(window=W, min_periods=W)
    dt_seconds = (W - 1) * CONFIG["cadence_seconds"]
    for c in cols:
        out[f"{c}_mean_W{W}"] = rolling[c].mean()
        out[f"{c}_std_W{W}"]  = rolling[c].std(ddof=0)
        out[f"{c}_max_W{W}"]  = rolling[c].max()
        out[f"{c}_min_W{W}"]  = rolling[c].min()
        out[f"{c}_ramp_W{W}"] = df[c].diff(W - 1) / dt_seconds
    return out


def delta_features(df: pd.DataFrame, cols: list[str], *, lags: list[int]) -> pd.DataFrame:
    """Build lagged-difference features at each lag in `lags`."""
    out = pd.DataFrame(index=df.index)
    for c in cols:
        for L in lags:
            out[f"{c}_d{L}"] = df[c].diff(L)
    return out


def build_common_derived(common_df: pd.DataFrame) -> pd.DataFrame:
    """Return the canonical-18 base columns plus sliding-window and delta features."""
    base = common_df[CANONICAL_COLS]
    sw   = sliding_window_features(common_df, CANONICAL_COLS, W=CONFIG["window_W"])
    dl   = delta_features(common_df, CANONICAL_COLS, lags=CONFIG["delta_lags"])
    feats = pd.concat([base, sw, dl], axis=1)
    # Carry the partition column through for downstream slicing.
    if "partition" in common_df.columns:
        feats["partition"] = common_df["partition"].values
    return feats
```

```python
# Cell 4.2. Build common derived features for all three sites.
# These columns are identical across sites and are the inputs to the cross-site protocol.
# Expected output: feature-frame shapes and column-count confirmation.

_cordoba_features = build_common_derived(_cordoba_common)
_site_a_features  = build_common_derived(_site_a_common)
_site_b_features  = build_common_derived(_site_b_common)

# Strip the partition column when counting feature dimensionality.
def _feat_cols(df: pd.DataFrame) -> list[str]:
    return [c for c in df.columns if c != "partition"]

_n_features_common = len(_feat_cols(_cordoba_features))
print(f"Common derived feature dimensionality: {_n_features_common} columns "
      f"(expected = 18 base + 18*5 sliding + 18*3 delta = 162).")
print(f"Cordoba features shape : {_cordoba_features.shape}")
print(f"Site A features shape  : {_site_a_features.shape}")
print(f"Site B features shape  : {_site_b_features.shape}")
```

```python
# Cell 4.3. Cordoba spectrum-shape features.
# R_oe, R_tri, C_U computed per phase for both voltage and current using orders 2 to 11.
# Expected output: shape of the spectrum-shape frame.

_eps         = CONFIG["epsilon"]
_orders_all  = list(range(2, 12))                  # 2..11 inclusive
_orders_odd  = CONFIG["spectrum_orders"]["odd"]    # {3, 5, 7, 9, 11}
_orders_even = CONFIG["spectrum_orders"]["even"]   # {2, 4, 6, 8, 10}
_orders_tri  = CONFIG["spectrum_orders"]["triplen"]


def spectrum_shape_cordoba(
    grid_df: pd.DataFrame,
    *,
    voltage_phases: list[str],
    current_phases: list[str],
) -> pd.DataFrame:
    """Compute R_oe, R_tri, C_U for Cordoba voltage and current per phase."""
    out = pd.DataFrame(index=grid_df.index)

    def _ratio_block(prefix_template: str, phase_list: list[str], tag: str) -> None:
        for ph in phase_list:
            mags2 = {
                h: grid_df[prefix_template.format(order=h, phase=ph)].astype(float) ** 2
                for h in _orders_all
            }
            sum_odd  = sum(mags2[h] for h in _orders_odd  if h in mags2)
            sum_even = sum(mags2[h] for h in _orders_even if h in mags2)
            sum_all  = sum(mags2.values())
            sum_tri  = sum(mags2[h] for h in _orders_tri  if h in mags2)
            sum_h_w  = sum(h * mags2[h] for h in mags2)
            out[f"R_oe_{tag}_{ph}"]  = sum_odd / (sum_even + _eps)
            out[f"R_tri_{tag}_{ph}"] = sum_tri / (sum_all  + _eps)
            out[f"C_U_{tag}_{ph}"]   = sum_h_w / (sum_all  + _eps)

    _ratio_block("vmag{order}{phase}", voltage_phases, tag="v")
    _ratio_block("imag{order}{phase}", current_phases, tag="i")
    return out


_cordoba_spectrum = spectrum_shape_cordoba(
    _cordoba_grid,
    voltage_phases=CORDOBA_VOLTAGE_PHASES,
    current_phases=CORDOBA_CURRENT_PHASES,
)
print(f"Cordoba spectrum-shape features shape: {_cordoba_spectrum.shape} "
      f"(expected cols = 3 metrics x ({len(CORDOBA_VOLTAGE_PHASES)} V + {len(CORDOBA_CURRENT_PHASES)} I) "
      f"= {3 * (len(CORDOBA_VOLTAGE_PHASES) + len(CORDOBA_CURRENT_PHASES))}).")
```

```python
# Cell 4.4. Cordoba complex-phasor features.
# Per-order P, Q, complex impedance Re/Im, circular phase embedding, and pr ratios at target orders.
# Expected output: shape of the phasor-feature frame and a head sample.

_target_orders = CONFIG["target_orders_cordoba"]    # [3, 5, 7, 9, 11]


def cordoba_phasor_features(
    grid_df: pd.DataFrame,
    *,
    voltage_phases: list[str],
    target_orders: list[int],
) -> pd.DataFrame:
    """Build harmonic power, complex impedance, circular embedding, and pr features.

    Phases here are the voltage-side phases a/b/c (current uses the same phase
    labels for paired V and I quantities). The neutral current channel is not
    paired with a voltage and is omitted from impedance and harmonic-power features.
    """
    out = pd.DataFrame(index=grid_df.index)
    eps = _eps
    for ph in voltage_phases:
        for h in target_orders:
            v_mag = grid_df[f"vmag{h}{ph}"].astype(float).to_numpy()
            v_ph  = np.deg2rad(grid_df[f"vph{h}{ph}"].astype(float).to_numpy())
            i_mag = grid_df[f"imag{h}{ph}"].astype(float).to_numpy()
            i_ph  = np.deg2rad(grid_df[f"iph{h}{ph}"].astype(float).to_numpy())

            U = v_mag * np.exp(1j * v_ph)
            Iv = i_mag * np.exp(1j * i_ph)

            # Harmonic complex power S_h = U_h * conj(I_h); P_h = Re, Q_h = Im.
            S_h = U * np.conj(Iv)
            out[f"P_h{h}_{ph}"] = S_h.real
            out[f"Q_h{h}_{ph}"] = S_h.imag

            # Complex impedance Z_h = U_h / I_h with NaN guard on near-zero current magnitude.
            denom_safe = np.where(np.abs(Iv) < eps, np.nan, Iv)
            Z = U / denom_safe
            out[f"Zr_h{h}_{ph}"] = Z.real
            out[f"Zi_h{h}_{ph}"] = Z.imag

            # Circular phase embedding (sin, cos in radians).
            out[f"vph_sin_h{h}_{ph}"] = np.sin(v_ph)
            out[f"vph_cos_h{h}_{ph}"] = np.cos(v_ph)
            out[f"iph_sin_h{h}_{ph}"] = np.sin(i_ph)
            out[f"iph_cos_h{h}_{ph}"] = np.cos(i_ph)

            # Prevailing-phasor confidence ratios pass through as features.
            out[f"vpr_h{h}_{ph}"] = grid_df[f"vpr{h}{ph}"].astype(float).to_numpy()
            out[f"ipr_h{h}_{ph}"] = grid_df[f"ipr{h}{ph}"].astype(float).to_numpy()
    return out


_cordoba_phasor = cordoba_phasor_features(
    _cordoba_grid,
    voltage_phases=CORDOBA_VOLTAGE_PHASES,
    target_orders=_target_orders,
)
_n_phasor_per_phase = 10 * len(_target_orders)
print(f"Cordoba phasor features shape: {_cordoba_phasor.shape} "
      f"(expected cols = {_n_phasor_per_phase} per phase x {len(CORDOBA_VOLTAGE_PHASES)} phases "
      f"= {_n_phasor_per_phase * len(CORDOBA_VOLTAGE_PHASES)}).")
```

```python
# Cell 4.5. Assemble Cordoba enriched feature frame.
# Common derived features + spectrum-shape + phasor features + circular embedding + pr ratios.
# Expected output: shape of the enriched frame.

_cordoba_features_enriched = pd.concat(
    [
        _cordoba_features.drop(columns=["partition"]),
        _cordoba_spectrum,
        _cordoba_phasor,
    ],
    axis=1,
)
_cordoba_features_enriched["partition"] = _cordoba_features["partition"].values

print(f"Cordoba enriched features shape: {_cordoba_features_enriched.shape}")
print(f"  Common derived  : {len(_feat_cols(_cordoba_features))} cols")
print(f"  Spectrum-shape  : {_cordoba_spectrum.shape[1]} cols")
print(f"  Phasor / pr     : {_cordoba_phasor.shape[1]} cols")
print(f"  Total           : {len(_feat_cols(_cordoba_features_enriched))} cols (excluding 'partition').")
```

```python
# Cell 4.6. Fit per-site scalers on training partitions.
# Three scalers: cordoba_common, cordoba_enriched, site_a_common, site_b_common.
# Cordoba-frozen common scaler (Module 3) is separate and pinned for cross-site transfer.
# Expected output: scaler artefact paths and per-feature fit-row counts printed.


def fit_save_scaler(
    feature_df: pd.DataFrame,
    *,
    scaler_name: str,
    site_name: str,
) -> Path:
    """Fit a StandardScaler on training rows of `feature_df` and pickle it.

    NaN rows (typically the first W-1 rows of sliding-window features) are dropped
    before the fit. The pickled artefact records the feature column list, the
    number of fit rows, and the partition used.
    """
    train_rows = feature_df.loc[feature_df["partition"].eq("train")]
    feat_cols  = [c for c in feature_df.columns if c != "partition"]
    train_clean = train_rows[feat_cols].dropna()
    scaler = StandardScaler().fit(train_clean.values)
    out_path = MODEL_DIR / f"scaler_{scaler_name}.joblib"
    joblib.dump(
        {
            "scaler":       scaler,
            "feature_cols": feat_cols,
            "fit_n":        int(train_clean.shape[0]),
            "fit_site":     site_name,
            "fit_partition":"train",
        },
        out_path,
    )
    print(f"  Fit scaler '{scaler_name}': n={train_clean.shape[0]} rows, "
          f"d={len(feat_cols)} features -> {out_path}")
    return out_path


print("Fitting per-site scalers on training partitions only:")
fit_save_scaler(_cordoba_features,           scaler_name="cordoba_common",   site_name="cordoba")
fit_save_scaler(_cordoba_features_enriched,  scaler_name="cordoba_enriched", site_name="cordoba")
fit_save_scaler(_site_a_features,            scaler_name="site_a_common",    site_name="site_a")
fit_save_scaler(_site_b_features,            scaler_name="site_b_common",    site_name="site_b")
```

```python
# Cell 4.7. Cache the four feature frames to parquet.
# Expected output: file paths and row x column counts.

_cordoba_features.to_parquet(INTERIM_DIR / "cordoba_features.parquet")
_cordoba_features_enriched.to_parquet(INTERIM_DIR / "cordoba_features_enriched.parquet")
_site_a_features.to_parquet(INTERIM_DIR / "site_a_features.parquet")
_site_b_features.to_parquet(INTERIM_DIR / "site_b_features.parquet")

_cache_table = pd.DataFrame(
    {
        "frame": ["cordoba_common", "cordoba_enriched", "site_a_common", "site_b_common"],
        "rows":  [_cordoba_features.shape[0], _cordoba_features_enriched.shape[0],
                  _site_a_features.shape[0], _site_b_features.shape[0]],
        "cols":  [len(_feat_cols(_cordoba_features)),
                  len(_feat_cols(_cordoba_features_enriched)),
                  len(_feat_cols(_site_a_features)),
                  len(_feat_cols(_site_b_features))],
    }
)
print(_cache_table.to_string(index=False))
```

## Module 5. Detection state machine with hysteresis and dwell

This module turns the per-sample threshold-cross labels produced in Modules 2 and 3 into operationally meaningful alerts. A naive instantaneous comparison of UTHD against the IEEE 519-2022 watch and exceedance thresholds is excessively chatty in the presence of measurement noise and short transient excursions. Two physically motivated filters are applied:

- **Hysteresis** with band $\Delta_h$ in percentage points. To upgrade from a lower state, UTHD must exceed the upper threshold by $+\Delta_h$. To downgrade, UTHD must fall below the lower threshold by $-\Delta_h$. The default $\Delta_h = 0.2$pp matches typical PQ analyser repeatability.
- **Dwell time** $D_{\min}$ in 10-minute reporting intervals. A candidate state must be observed for $D_{\min}$ consecutive intervals before being confirmed. The default $D_{\min} = 2$ corresponds to the IEC 61000-4-30 short-time aggregation principle (a transient that does not survive a full short-time block does not constitute a compliance event).

The ordinal mapping is fixed: Normal $= 0$, Watch $= 1$, Exceedance $= 2$. Worst-state aggregation across phases uses the maximum ordinal at each timestamp.

In addition to confirmed-state assignment, this module performs a $\tau_W$ sensitivity sweep over $\{4.0, 5.0, 6.4\}\%$ to characterise how the watch-class population responds to threshold relaxation, since the IEC-recommended weekly-95th-percentile aggregation produces sparse exceedance events on quasi-stationary feeders. The sweep gives the empirical justification for retaining $\tau_W = 6.4\%$ (i.e. $0.8 \cdot \tau_E$) as the operational watch threshold while documenting that detection rate does not vanish under reasonable threshold perturbation.

```python
# Cell 5.1. Hysteresis + dwell state machine.
# Operates on a single UTHD time series. Returns confirmed states as int ordinals.

def apply_hysteresis_dwell(
    uthd: np.ndarray,
    tau_W: float,
    tau_E: float,
    delta_h: float,
    d_min: int,
    initial_state: int = 0,
) -> np.ndarray:
    """Confirm states from a raw UTHD series using hysteresis + dwell time.

    Hysteresis: upgrade requires UTHD strictly greater than threshold + delta_h;
    downgrade requires UTHD strictly less than threshold - delta_h.

    Dwell: a candidate state different from the current confirmed state must
    persist for d_min consecutive samples before becoming confirmed.

    Parameters
    ----------
    uthd : 1-D float array of UTHD samples in percent.
    tau_W : Watch threshold in percent.
    tau_E : Exceedance threshold in percent.
    delta_h : Hysteresis band in percent.
    d_min : Minimum number of consecutive samples to confirm a transition.
    initial_state : Starting ordinal state. Defaults to Normal (0).

    Returns
    -------
    Array of int ordinals, same length as uthd.
    """
    n = len(uthd)
    confirmed = np.empty(n, dtype=int)
    current = int(initial_state)
    pending_state = current
    pending_count = 0

    for t in range(n):
        x = float(uthd[t])

        if current == 2:
            target = 2 if x >= tau_E - delta_h else (1 if x >= tau_W - delta_h else 0)
        elif current == 1:
            if x >= tau_E + delta_h:
                target = 2
            elif x <= tau_W - delta_h:
                target = 0
            else:
                target = 1
        else:  # current == 0
            if x >= tau_E + delta_h:
                target = 2
            elif x >= tau_W + delta_h:
                target = 1
            else:
                target = 0

        if target == current:
            pending_state = current
            pending_count = 0
        else:
            if target == pending_state:
                pending_count += 1
            else:
                pending_state = target
                pending_count = 1
            if pending_count >= d_min:
                current = pending_state
                pending_count = 0

        confirmed[t] = current

    return confirmed
```

```python
# Cell 5.2. Apply the state machine per UTHD channel and aggregate to a worst-state series.
# Uses the categorical->object cast inside aggregate_worst_state (M2) to avoid pandas dtype errors.

def confirm_states_per_channel(
    df: pd.DataFrame,
    uthd_cols: list[str],
    state_keys: list[str],
    tau_W: float,
    tau_E: float,
    delta_h: float,
    d_min: int,
) -> pd.DataFrame:
    """Apply hysteresis+dwell to each UTHD channel and produce confirmed-state columns.

    Adds `state_conf_{key}` ordinal-state strings per channel and a
    `state_conf_worst` column holding the worst-case across channels.
    """
    out = df.copy()
    for col, key in zip(uthd_cols, state_keys):
        ords = apply_hysteresis_dwell(
            uthd=out[col].to_numpy(dtype=float, na_value=0.0),
            tau_W=tau_W, tau_E=tau_E, delta_h=delta_h, d_min=d_min,
        )
        out[f"state_conf_{key}"] = pd.Series(ords, index=out.index).map(ORD_TO_STATE).astype("string")
    out["state_conf_worst"] = aggregate_worst_state(
        out[[f"state_conf_{k}" for k in state_keys]]
    )
    return out
```

```python
# Cell 5.5. Detection metrics. Compares confirmed states to raw threshold-cross states.
# Treats raw exceedance as the ground-truth-positive class for MER.
# Treats raw normal as the negative class for FAR.

def detection_metrics(raw_state: pd.Series, conf_state: pd.Series) -> dict:
    """Confusion matrix and rate metrics relating raw and confirmed state series."""
    rs = raw_state.map(STATE_TO_ORD).astype(int).to_numpy()
    cs = conf_state.map(STATE_TO_ORD).astype(int).to_numpy()
    cm = np.zeros((3, 3), dtype=int)
    for r_o, c_o in zip(rs, cs):
        cm[r_o, c_o] += 1
    raw_normal_total = int(cm[0].sum())
    far = float((cm[0, 1] + cm[0, 2]) / max(raw_normal_total, 1))
    raw_exc_total = int(cm[2].sum())
    mer = float(cm[2, 0] / max(raw_exc_total, 1)) if raw_exc_total > 0 else float("nan")
    raw_watch_total = int(cm[1].sum())
    suppressed_watch_rate = (
        float(cm[1, 0] / max(raw_watch_total, 1)) if raw_watch_total > 0 else float("nan")
    )
    return {
        "confusion_matrix": cm,
        "false_alarm_rate": far,
        "missed_exceedance_rate": mer,
        "suppressed_watch_rate": suppressed_watch_rate,
        "raw_normal_count": raw_normal_total,
        "raw_watch_count": raw_watch_total,
        "raw_exceedance_count": raw_exc_total,
    }
```

```python
# Cell 5.6. Compute detection metrics on each site and write a results table.

_metrics_records = []
for _site, _frame in [
    ("Cordoba", _cordoba_grid_with_conf),
    ("Site_A",  _site_a_grid_with_conf),
    ("Site_B",  _site_b_grid_with_conf),
]:
    _m = detection_metrics(_frame["state_worst"], _frame["state_conf_worst"])
    _metrics_records.append({
        "site": _site,
        "raw_normal_n": _m["raw_normal_count"],
        "raw_watch_n":  _m["raw_watch_count"],
        "raw_exceed_n": _m["raw_exceedance_count"],
        "false_alarm_rate":      round(_m["false_alarm_rate"], 6),
        "suppressed_watch_rate": round(_m["suppressed_watch_rate"], 6)
            if not np.isnan(_m["suppressed_watch_rate"]) else float("nan"),
        "missed_exceedance_rate": _m["missed_exceedance_rate"],
    })

_table_m5_metrics = pd.DataFrame.from_records(_metrics_records)
_path = TBL_DIR / "table_m5_detection_metrics.csv"
_table_m5_metrics.to_csv(_path, index=False)
print(f"Saved: {_path}")
print(_table_m5_metrics.to_string(index=False))
```

```python
# Cell 5.7. Mean lead time from first confirmed Watch to subsequent first confirmed Exceedance.
# An "event" begins at the Normal->Watch transition and ends when the series returns to Normal.
# Lead time is undefined if the event never reaches Exceedance. Reported in 10-minute intervals.

def event_lead_times(conf_ordinals: np.ndarray) -> list[int]:
    """Return the per-event lead time (number of intervals from Watch entry to first Exceedance)."""
    leads: list[int] = []
    in_event = False
    watch_entry_idx = -1
    for t, s in enumerate(conf_ordinals):
        if not in_event and s >= 1:
            in_event = True
            watch_entry_idx = t
        elif in_event and s == 0:
            in_event = False
            watch_entry_idx = -1
        elif in_event and s == 2 and watch_entry_idx >= 0:
            leads.append(t - watch_entry_idx)
            watch_entry_idx = -1  # only record the first exceedance per event
    return leads


_lead_records = []
for _site, _frame in [
    ("Cordoba", _cordoba_grid_with_conf),
    ("Site_A",  _site_a_grid_with_conf),
    ("Site_B",  _site_b_grid_with_conf),
]:
    _ords = _frame["state_conf_worst"].map(STATE_TO_ORD).astype(int).to_numpy()
    _leads = event_lead_times(_ords)
    _lead_records.append({
        "site": _site,
        "n_events_with_exceedance": len(_leads),
        "mean_lead_time_intervals": float(np.mean(_leads)) if _leads else float("nan"),
        "median_lead_time_intervals": float(np.median(_leads)) if _leads else float("nan"),
    })

_table_m5_lead = pd.DataFrame.from_records(_lead_records)
_path = TBL_DIR / "table_m5_lead_time.csv"
_table_m5_lead.to_csv(_path, index=False)
print(f"Saved: {_path}")
print(_table_m5_lead.to_string(index=False))
```

```python
# Cell 5.8. tau_W sensitivity sweep on Cordoba. Sweep tau_W in {4.0, 5.0, 6.4}%, hold tau_E fixed.
# Records both raw threshold-cross counts and confirmed-state counts at each setting.

def relabel_with_tau(
    df: pd.DataFrame, uthd_cols: list[str], state_keys: list[str],
    tau_W: float, tau_E: float,
) -> pd.Series:
    """Recompute the worst-state ternary label using a given tau_W."""
    sub = pd.DataFrame(index=df.index)
    for col, key in zip(uthd_cols, state_keys):
        sub[f"state_{key}"] = label_uthd(df[col], tau_W=tau_W, tau_E=tau_E)
    return aggregate_worst_state(sub)


_sweep_records = []
for _tw in CONFIG["tau_W_sweep_pct"]:
    _raw_worst = relabel_with_tau(
        _cordoba_grid, _cordoba_uthd_cols, CORDOBA_VOLTAGE_PHASES,
        tau_W=_tw, tau_E=CONFIG["thresholds"]["tau_E_pct"],
    )
    _conf_frame = confirm_states_per_channel(
        df=_cordoba_grid,
        uthd_cols=_cordoba_uthd_cols,
        state_keys=CORDOBA_VOLTAGE_PHASES,
        tau_W=_tw,
        tau_E=CONFIG["thresholds"]["tau_E_pct"],
        delta_h=CONFIG["thresholds"]["delta_h_pct"],
        d_min=CONFIG["thresholds"]["dwell_min_intervals"],
    )
    _conf_worst = _conf_frame["state_conf_worst"]
    _raw_counts = _raw_worst.value_counts(dropna=False)
    _conf_counts = _conf_worst.value_counts(dropna=False)
    _sweep_records.append({
        "tau_W_pct": _tw,
        "raw_Normal":      int(_raw_counts.get("Normal", 0)),
        "raw_Watch":       int(_raw_counts.get("Watch", 0)),
        "raw_Exceedance":  int(_raw_counts.get("Exceedance", 0)),
        "conf_Normal":     int(_conf_counts.get("Normal", 0)),
        "conf_Watch":      int(_conf_counts.get("Watch", 0)),
        "conf_Exceedance": int(_conf_counts.get("Exceedance", 0)),
    })

_table_m5_sweep = pd.DataFrame.from_records(_sweep_records)
_path = TBL_DIR / "table_m5_tau_w_sweep.csv"
_table_m5_sweep.to_csv(_path, index=False)
print(f"Saved: {_path}")
print(_table_m5_sweep.to_string(index=False))
```

## Module 6. Multi-horizon HGBR forecasting with blocked walk-forward cross-validation

This module fits the proposed model class — `HistGradientBoostingRegressor` (HGBR), a histogram-binned gradient-boosted tree regressor — to the engineered telemetry features built in Module 4. Three design choices are baked in here and carried forward to Modules 7 (conformal calibration), 8 (baselines), and 9 (evaluation):

- **Multi-horizon as independent regressors.** A separate model is fitted per forecast horizon $h \in \{1, 3, 6, 12\}$ ten-minute steps, equivalent to $\{10, 30, 60, 120\}$ minutes ahead. This avoids horizon-coupled error propagation that would arise from recursive single-step rollouts.
- **Per-target single-output regression.** HGBR is natively single-output. Multiple targets (per-phase UTHD on every site; plus per-phase per-order voltage harmonic magnitudes on Cordoba) are fitted as independent regressors. This sacrifices joint distribution structure across targets in exchange for hyperparameter parsimony and reproducibility.
- **Hyperparameter selection via blocked walk-forward cross-validation.** The training partition is split into $K = 5$ chronologically contiguous blocks. For each candidate hyperparameter configuration, the regressor is fit on the union of blocks $\{1, \dots, k\}$ and scored on block $k+1$ for $k \in \{1, \dots, K-1\}$. The grid is searched once, on a primary $(\text{target}, \text{horizon})$ pair on Cordoba (UTHD phase B at $h=1$ step). The best configuration is reused for every other target/horizon combination on the same dataset, on grounds of HGBR's well-documented hyperparameter robustness across structurally similar regression tasks. This is a deliberate scope-trimming decision and is documented in the thesis methodology.

The hyperparameter search space, locked from Chapter 3:

- `max_iter` $\in \{200, 500, 1000\}$
- `learning_rate` $\in \{0.01, 0.05, 0.1\}$
- `max_depth` $\in \{\text{None}, 5, 8, 12\}$
- `l2_regularization` $\in \{0.0, 0.1, 1.0\}$
- `max_bins` fixed at 255 (default)
- Early stopping enabled with `n_iter_no_change` $= 30$, `tol = 10^{-7}$

Total grid size is $3 \times 3 \times 4 \times 3 = 108$ configurations. Combined with $K-1 = 4$ scoring folds, this is 432 fits at the primary target/horizon. Scoring metric is mean absolute error (MAE) on the held-out block.

```python
# Cell 6.1. HGBR factory with the critical nan->None fix for max_depth.
# When hyperparameter rows survive a pandas DataFrame round-trip (e.g. for logging),
# Python `None` becomes `NaN`. scikit-learn's HGBR raises if max_depth is NaN.

from sklearn.ensemble import HistGradientBoostingRegressor


def _coerce_depth(d) -> int | None:
    """Convert a possibly-NaN max_depth back to None or int."""
    if d is None:
        return None
    if isinstance(d, float) and math.isnan(d):
        return None
    return int(d)


def make_hgbr(
    max_iter: int,
    learning_rate: float,
    max_depth,
    l2_regularization: float,
    seed: int = SEED,
) -> HistGradientBoostingRegressor:
    """Build an HGBR with locked early-stopping configuration."""
    return HistGradientBoostingRegressor(
        loss="absolute_error",
        max_iter=int(max_iter),
        learning_rate=float(learning_rate),
        max_depth=_coerce_depth(max_depth),
        l2_regularization=float(l2_regularization),
        max_bins=255,
        early_stopping=True,
        n_iter_no_change=30,
        tol=1e-7,
        validation_fraction=0.1,
        random_state=seed,
    )
```

```python
# Cell 6.2. Build supervised samples (X, y) at a given forecast horizon.
# A target column at sample t is the value of the source column at sample t+h.
# Last h rows are dropped because their target falls outside the observation window.

def build_supervised_at_horizon(
    feature_frame: pd.DataFrame,
    source_frame: pd.DataFrame,
    target_col: str,
    horizon_steps: int,
) -> tuple[pd.DataFrame, pd.Series, pd.Series]:
    """Return (X, y, partition) aligned for forecasting `target_col` at horizon `h`.

    `feature_frame` must carry a `partition` column. `source_frame` must
    contain `target_col` and share an index with `feature_frame`.
    """
    if target_col not in source_frame.columns:
        raise KeyError(f"target column '{target_col}' missing from source frame")
    common_idx = feature_frame.index.intersection(source_frame.index)
    feat = feature_frame.loc[common_idx].copy()
    src = source_frame.loc[common_idx, target_col].astype(float)
    y_future = src.shift(-horizon_steps)
    valid_mask = y_future.notna()
    feat = feat.loc[valid_mask]
    y = y_future.loc[valid_mask]
    partition = feat["partition"].astype("string")
    X = feat.drop(columns=["partition"], errors="ignore")
    return X, y, partition
```

```python
# Cell 6.3. Blocked walk-forward CV split generator. K contiguous blocks, K-1 scoring folds.
# Fold k trains on blocks {1..k} and scores on block k+1.

def blocked_walk_forward_splits(n_samples: int, k_blocks: int) -> list[tuple[np.ndarray, np.ndarray]]:
    """Yield (train_idx, test_idx) pairs for blocked walk-forward CV."""
    if k_blocks < 2:
        raise ValueError("k_blocks must be at least 2")
    edges = np.linspace(0, n_samples, k_blocks + 1, dtype=int)
    folds: list[tuple[np.ndarray, np.ndarray]] = []
    for k in range(1, k_blocks):
        train_idx = np.arange(0, edges[k])
        test_idx  = np.arange(edges[k], edges[k + 1])
        if len(train_idx) > 0 and len(test_idx) > 0:
            folds.append((train_idx, test_idx))
    return folds
```

```python
# Cell 6.4. Score one hyperparameter configuration on the training partition.
# Returns mean and per-fold MAE across walk-forward folds.

from sklearn.metrics import mean_absolute_error


def score_hgbr_config(
    X_train: pd.DataFrame,
    y_train: pd.Series,
    config: dict,
    k_blocks: int,
    seed: int = SEED,
) -> dict:
    """Run blocked walk-forward CV on a single HGBR config. Returns aggregate MAE."""
    X = X_train.to_numpy(dtype=float)
    y = y_train.to_numpy(dtype=float)
    folds = blocked_walk_forward_splits(len(X), k_blocks)
    fold_maes: list[float] = []
    for train_idx, test_idx in folds:
        model = make_hgbr(seed=seed, **config)
        model.fit(X[train_idx], y[train_idx])
        preds = model.predict(X[test_idx])
        fold_maes.append(float(mean_absolute_error(y[test_idx], preds)))
    return {
        "mean_mae": float(np.mean(fold_maes)),
        "std_mae":  float(np.std(fold_maes)),
        "fold_maes": fold_maes,
    }
```

```python
# Cell 6.5. Run the full HP grid on the primary CV target (Cordoba UTHD phase B at h=1 step).
# This is the heaviest cell in the notebook. Expect tens of minutes on commodity hardware.
# Targets and horizons are reused for the per-target final-fit step in Cell 6.7.

PRIMARY_TARGET = "thdvb"
PRIMARY_HORIZON_STEPS = 1   # 10-minute look-ahead

_X_full, _y_full, _partition = build_supervised_at_horizon(
    feature_frame=_cordoba_features_enriched,
    source_frame=_cordoba_grid,
    target_col=PRIMARY_TARGET,
    horizon_steps=PRIMARY_HORIZON_STEPS,
)
_train_mask = (_partition == "train")
_X_train_primary = _X_full.loc[_train_mask]
_y_train_primary = _y_full.loc[_train_mask]
print(f"Training samples for primary CV: {len(_X_train_primary):,}")
print(f"Feature dimensionality: {_X_train_primary.shape[1]}")

_grid = []
for _mi in CONFIG["hgbr_grid"]["max_iter"]:
    for _lr in CONFIG["hgbr_grid"]["learning_rate"]:
        for _md in CONFIG["hgbr_grid"]["max_depth"]:
            for _l2 in CONFIG["hgbr_grid"]["l2_regularization"]:
                _grid.append({
                    "max_iter": _mi,
                    "learning_rate": _lr,
                    "max_depth": _md,
                    "l2_regularization": _l2,
                })
print(f"Hyperparameter grid size: {len(_grid)}")

_results = []
for _i, _cfg in enumerate(_grid, start=1):
    _t0 = time.time()
    _r = score_hgbr_config(
        _X_train_primary, _y_train_primary, _cfg,
        k_blocks=CONFIG["cv_k_blocks"], seed=SEED,
    )
    _results.append({**_cfg, **_r, "wall_time_s": round(time.time() - _t0, 2)})
    if _i % 12 == 0 or _i == len(_grid):
        print(f"  [{_i:>3}/{len(_grid)}] last MAE = {_r['mean_mae']:.4f}  "
              f"({_results[-1]['wall_time_s']}s)")

_cv_results_df = pd.DataFrame.from_records(_results).drop(columns=["fold_maes"])
_cv_results_df = _cv_results_df.sort_values("mean_mae", ascending=True).reset_index(drop=True)
_path = TBL_DIR / "table_m6_cv_results.csv"
_cv_results_df.to_csv(_path, index=False)
print(f"Saved: {_path}")
print(_cv_results_df.head(10).to_string(index=False))
```

```python
# Cell 6.6. Lock the best configuration and persist it as a JSON manifest for downstream modules.

_best_row = _cv_results_df.iloc[0]
BEST_HGBR_CONFIG = {
    "max_iter":          int(_best_row["max_iter"]),
    "learning_rate":     float(_best_row["learning_rate"]),
    "max_depth":         _coerce_depth(_best_row["max_depth"]),
    "l2_regularization": float(_best_row["l2_regularization"]),
}
print("Best HGBR configuration (lowest mean walk-forward CV MAE):")
print(json.dumps(BEST_HGBR_CONFIG, indent=2))
print(f"  CV mean MAE = {_best_row['mean_mae']:.4f} (std {_best_row['std_mae']:.4f})")

_manifest = {
    "primary_target": PRIMARY_TARGET,
    "primary_horizon_steps": PRIMARY_HORIZON_STEPS,
    "feature_set": "cordoba_enriched",
    "k_blocks": CONFIG["cv_k_blocks"],
    "best_config": BEST_HGBR_CONFIG,
    "cv_mean_mae": float(_best_row["mean_mae"]),
    "cv_std_mae":  float(_best_row["std_mae"]),
}
with open(MODEL_DIR / "best_hgbr_config.json", "w") as _fh:
    json.dump(_manifest, _fh, indent=2)
print(f"Saved: {MODEL_DIR / 'best_hgbr_config.json'}")
```

```python
# Cell 6.7. Fit one HGBR per (target, horizon) on Cordoba train+val using BEST_HGBR_CONFIG.
# Models are persisted under outputs/models/. The validation partition is held out from the
# fit set in this cell (used only for conformal calibration in Module 7).

_cordoba_targets = [
    "thdva", "thdvb", "thdvc",
    *[f"vmag{h}{p}" for h in [3, 5, 7, 9, 11] for p in CORDOBA_VOLTAGE_PHASES],
]
_horizons = CONFIG["forecast_horizons_steps"]

print(f"Fitting {len(_cordoba_targets)} targets x {len(_horizons)} horizons "
      f"= {len(_cordoba_targets) * len(_horizons)} models on Cordoba (enriched).")

_fit_records = []
for _target in _cordoba_targets:
    if _target not in _cordoba_grid.columns:
        print(f"  [skip] {_target} not in Cordoba grid")
        continue
    for _h in _horizons:
        _Xh, _yh, _ph = build_supervised_at_horizon(
            feature_frame=_cordoba_features_enriched,
            source_frame=_cordoba_grid,
            target_col=_target,
            horizon_steps=_h,
        )
        _train_mask_h = (_ph == "train")
        _Xtr = _Xh.loc[_train_mask_h].to_numpy(dtype=float)
        _ytr = _yh.loc[_train_mask_h].to_numpy(dtype=float)
        if len(_Xtr) == 0:
            continue
        _t0 = time.time()
        _model = make_hgbr(seed=SEED, **BEST_HGBR_CONFIG)
        _model.fit(_Xtr, _ytr)
        _wall = round(time.time() - _t0, 2)
        _path = MODEL_DIR / f"hgbr_cordoba_enriched_{_target}_h{_h}.joblib"
        joblib.dump(_model, _path)
        _fit_records.append({
            "feature_set": "cordoba_enriched",
            "target": _target,
            "horizon_steps": _h,
            "n_train": len(_Xtr),
            "wall_time_s": _wall,
            "artefact": _path.name,
        })

_fit_table = pd.DataFrame.from_records(_fit_records)
_path = TBL_DIR / "table_m6_fitted_models.csv"
_fit_table.to_csv(_path, index=False)
print(f"Saved: {_path}")
print(f"Total models fitted: {len(_fit_table)}")
print(_fit_table.head(8).to_string(index=False))
```

```python
# Cell 6.8. Fit the Cordoba common-features-only model (3 UTHD targets x 4 horizons).
# Trained on cordoba_features (162-col common derived). This artefact is the basis of the
# cross-site transfer evaluation in Module 10. It is intentionally separate from the
# enriched-feature model to keep the transfer protocol scoped to the canonical feature set.

_uthd_targets = ["thdva", "thdvb", "thdvc"]
_transfer_records = []
for _target in _uthd_targets:
    for _h in _horizons:
        _Xh, _yh, _ph = build_supervised_at_horizon(
            feature_frame=_cordoba_features,
            source_frame=_cordoba_grid,
            target_col=_target,
            horizon_steps=_h,
        )
        _Xtr = _Xh.loc[_ph == "train"].to_numpy(dtype=float)
        _ytr = _yh.loc[_ph == "train"].to_numpy(dtype=float)
        _t0 = time.time()
        _model = make_hgbr(seed=SEED, **BEST_HGBR_CONFIG)
        _model.fit(_Xtr, _ytr)
        _wall = round(time.time() - _t0, 2)
        _path = MODEL_DIR / f"hgbr_cordoba_common_{_target}_h{_h}.joblib"
        joblib.dump(_model, _path)
        _transfer_records.append({
            "feature_set": "cordoba_common",
            "target": _target,
            "horizon_steps": _h,
            "n_train": len(_Xtr),
            "wall_time_s": _wall,
            "artefact": _path.name,
        })

_transfer_table = pd.DataFrame.from_records(_transfer_records)
_path = TBL_DIR / "table_m6_fitted_models_common.csv"
_transfer_table.to_csv(_path, index=False)
print(f"Saved: {_path}")
print(_transfer_table.to_string(index=False))
```
